```python
print(df['production.period'].unique())
print(len(df['production.period'].unique()))
print()
print(df['production.date.start'].unique())
print(len(df['production.date.start'].unique()))
print()
print(df['production.date.end'].unique())
print(len(df['production.date.end'].unique()))
print()

import numpy as np
# Turn date strings into integers
strStartDates = list(df['production.date.start'])
startDates = []
for date in strStartDates:
    #print(type(date))
    if type(date) is str:
        if date == '1902-07':
            intDate = 1902
        else:
            intDate = int(date)
        startDates += [intDate]
print("Earliest and Latest Start Dates")
print(np.min(startDates))
print(np.max(startDates))
print()
strEndDates = list(df['production.date.end'])
endDates = []
for date in strEndDates:
    if type(date) is str:
        if date == '1902-07':
            intDate = 1907
        else:
            intDate = int(date)
        endDates += [intDate]
print("Earliest and Latest End Dates")
print(np.min(endDates))
print(np.max(endDates))
```

One of the dates was a range, so I split the range into start date 1902 and end date 1907.

```python
sub_df = pd.DataFrame( { 'ID' : df.loc[:, 'DwD'], 'object_category' : df.loc[:, 'object_category'], 'object_name' : df.object_name, 'description' : df.description, 'creator' : df.creator, 'production_place' : df.loc[:, 'production.place'], 'production_date_start' : df.loc[:, 'production.date.start'], 'production_date_end' : df.loc[:, 'production.date.end'], 'current_location' : df.current_location })
sub_df.tail()
```

```python
sub_df.shape
```

```python
end_date_list = list(sub_df.production_date_end)
new_end_dates = []
#type(end_date_list[850])
#print(end_date_list[850])
for d in end_date_list:
    if type(d) == float:
        d = '0000'
    d = str(d)
    if d == '1902-07':
        d = '1907'
    new_end_dates += [d]
    
for new_d in new_end_dates:
    assert type(new_d) == str, "All new dates should be strings"
    assert len(new_d) == 4, "All new dates should be 4 digit years"
```

```python
start_date_list = list(sub_df.production_date_start)
new_start_dates = []
for sd in start_date_list:
    if type(sd) == float:
        sd = '0000'
    sd = str(sd)
    if sd == '1902-07':
        sd = '1902'
    new_start_dates += [sd]
    
for new_sd in new_start_dates:
    assert type(new_sd) == str, "All new dates should be strings"
    assert len(new_sd) == 4, "All new dates should be 4 digit years"
```

```python
assert len(new_end_dates) == len(new_start_dates), "There should be the same number of start and end dates"
```

```python
creator_list = list(sub_df.creator)
new_creators = []
for c in creator_list:
    if type(c) == float:
        c = 'Unknown'
    new_creators += [c]

for new_c in new_creators:
    assert type(new_c) == str, "All new creators should be strings"
```

```python
place_list = list(sub_df.production_place)
new_places = []
for p in place_list:
    if type(p) == float:
        p = 'Unknown'
    new_places += [p]
    
for new_p in new_places:
    assert type(new_p) == str, "All new dates should be strings"
```

```python
sub_df = pd.DataFrame( { 'ID' : df.loc[:, 'DwD'], 'object_category' : df.loc[:, 'object_category'], 'object_name' : df.object_name, 'description' : df.description, 'creator' : new_creators, 'production_place' : new_places, 'production_date_start' : new_start_dates, 'production_date_end' : new_end_dates, 'current_location' : df.current_location })
sub_df.head()
```

# Data Visualization Drafts

Data visualizations for 4 themes: creator, product date (start & end), current location, and production place

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# From https://seaborn.pydata.org/examples/color_palettes.html
#f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 6), sharex=True)

f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 20), sharex=False)

# Visualize the number of objects in each Hill House room
x1 = list(c_l_dict.keys())
y1 = list(c_l_dict.values())
g1 = sns.barplot(x1, y1, palette="Set1", ax=ax1)
ax1.set_xlabel("Room in Hill House")
g1.set_xticklabels("Room in Hill House", rotation=30)
ax1.set_ylabel("Number of Objects")

# Visualize the number of objects in each production place
x2 = list(p_p_dict.keys())
y2 = list(p_p_dict.values())
g2 = sns.barplot(x2, y2, palette="Set2", ax=ax2).set_xticklabels(rotation=30)
ax2.set_xlabel("Production Place")
g2.set_xticklabels("Production Place", rotation=30)
ax2.set_ylabel("Number of Objects")

# Visualize the number of objects made by each creator
x3 = list(c_dict.keys())
y3 = list(c_dict.values())
g3 = sns.barplot(x3, y3, palette="Set3", ax=ax3).set_xticklabels(rotation=30)
ax3.set_xlabel("Creator")
g3.set_xticklabels("Creator", rotation=30)
ax3.set_ylabel("Number of Objects")

# Visualize the number of objects each creator made
# Set appearance of plots
#sns.despine(bottom=True)
plt.title("Hill House Object Counts")
#plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=3)
plt.show()
```
